```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("heart attack.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data["age"].hist(bins=20)
plt.xlabel("Age")
```

```python
plt.figure(figsize=(6, 6))
sns.displot(x="age", data=data, hue="target", bins=30, color="BlGr")
```

```python
sns.pairplot(data[["age", "cp", "trestbps", 'chol', 'target']])
```

```python
plt.figure(figsize=(8, 8))
plt.bar(data.oldpeak, data.thalach, color=["grey", "black"]);
plt.xlabel("Oldpeak");
plt.ylabel("Thalach");
```

```python
corr = data.corr()
corr
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(corr,
            annot=True)
```
